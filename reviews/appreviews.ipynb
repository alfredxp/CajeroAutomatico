```python
#@title App Reviews from Google Store
from IPython.display import clear_output
try:
    from google_play_scraper import app, search, Sort, reviews, reviews_all
    from pprint import pprint
    import pandas as pd
    import numpy as np
    import json
    import pdb
except ModuleNotFoundError:
    %pip install google-play-scraper
    clear_output()
    %pip install google-search-scraper-python
    clear_output()
    %pip install pandas
    clear_output()
    
#static variables:
language = 'en'
countries = 'us'
appid = []

def searches(nameofSearch):
  result = search(
    nameofSearch,
    lang="en",  # defaults to 'en'
    country="us",  # defaults to 'us'
    n_hits=3  # defaults to 30 (= Google's maximum)
)
  return result

def getAppId(r):
  for d in r:
    id = (d['appId'])
    # appid.append(id)
    return id

def insertId(g, cont, rating, choice, leng):

  if choice == None:
    if rating == None:
      result, continuation_token = reviews(
      g,
      lang=leng, # defaults to 'en'
      country='us', # defaults to 'us'
      sort=Sort.NEWEST, # defaults to Sort.NEWEST
      count=cont # defaults to 100
      # defaults to None(means all score)
    )
      return result
    else:
      result, continuation_token = reviews(
      g,
      lang=leng, # defaults to 'en'
      country='us', # defaults to 'us'
      sort=Sort.NEWEST, # defaults to Sort.NEWEST
      count=cont, # defaults to 100
      filter_score_with=rating# defaults to None(means all score)
    )
      return result
 
  else:
    if rating == None:
      result = reviews_all(
      g,
      sleep_milliseconds=0, # defaults to 0
      lang=leng, # defaults to 'en'
      country='us', # defaults to 'us'
      # sort=Sort.MOST_RELEVANT, # defaults to Sort.MOST_RELEVANT
      filter_score_with=None
      )
      return result
    else: 
      result = reviews_all(
      g,
      sleep_milliseconds=0, # defaults to 0
      lang=leng, # defaults to 'en'
      country='us', # defaults to 'us'
      # sort=Sort.MOST_RELEVANT, # defaults to Sort.MOST_RELEVANT
      filter_score_with=rating # defaults to None(means all score)
      )
      return result

def googleplay(a, ct, r, c, l):
  searches1 = searches(a)
  appId = getAppId(searches1)
  reviews = insertId(appId, ct, r, c, l)
  return reviews

def googleplay1(a, ct, r, c, l):
  searches1 = searches(a)
  appId = getAppId(searches1)
  reviews = insertId(appId, ct, r, c, l)
  return reviews 

#@markdown * ### Input the name of the app: 
app = "sololearn" #@param {type:"string"}

#@markdown * ### All reviews?
choice = "Yes" #@param ["Yes", "No"]
if choice == "Yes":
  choice = "Yes"
  count = None
else:  
  choice = None
  #@markdown * ### How many reviews? (leave empty if you chose Yes to all reviews)
  count = 0 #@param {type:"number"}
#@markdown * ### Include specific rating?
cond = "No" #@param ["Yes", "No"]  
if cond == "Yes":
  #@markdown * ### Input the rating: (0-5) (leave empty if you chose No to specific rating)
  rating = 0 #@param {type:"integer"}
else:
  rating = None

#@markdown * ### Language:
language = "English" #@param ["English", "Spanish", "Both"]
if language == "English":
  lenguaje = 'en'
  googleplaydf = pd.DataFrame(np.array(googleplay(app, count, rating, choice, lenguaje)),columns=['review'])
  df5 = googleplaydf.join(pd.DataFrame(googleplaydf.pop('review').tolist()))

if language == "Spanish":
  lenguaje = 'es'
  googleplaydf = pd.DataFrame(np.array(googleplay(app, count, rating, choice, lenguaje)),columns=['review'])
  df5 = googleplaydf.join(pd.DataFrame(googleplaydf.pop('review').tolist()))
else: 
  lenguaje = 'en'
  googleplaydf = pd.DataFrame(np.array(googleplay(app, count, rating, choice, lenguaje)),columns=['review'])
  googleplaydf2 = googleplaydf.join(pd.DataFrame(googleplaydf.pop('review').tolist()))

  lenguaje = 'es'
  googleplaydf3 = pd.DataFrame(np.array(googleplay1(app, count, rating, choice, lenguaje)),columns=['review'])
  googleplaydf4 = googleplaydf3.join(pd.DataFrame(googleplaydf3.pop('review').tolist()))
  df5 = googleplaydf2.append(googleplaydf4, ignore_index=True)
  df5.sort_values(by='at')
  
#@markdown * ### Display data?
visualize = "Yes" #@param ["Yes", "No"]
if visualize == "Yes":
  display(df5)
else:
  print("No display")

#@markdown * ### Save to csv?
csv = "No" #@param ["Yes", "No"]
if csv == "Yes":
  df5.to_csv('googleplay-review.csv') 
else:
  print("No export")
```
